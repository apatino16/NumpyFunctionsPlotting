
# Basic Numpy Functions and Plotting    


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q1
Plot $y(x)=sin(2^x)$ for $x \in [-5, 8]$, with $x$ sampled in increments of
 $0.01$.

```python
x = np.arange(-5,8,0.01)
y = np.sin(2**x)

figure, axis = plt.subplots()

plt.plot(x,y)
plt.xlabel("x values from -5 to 8")
plt.ylabel("Sin(2**x)")
plt.title("Plot of sin(2**x) function from x=-5 to x=8")
plt.legend("sin(2**x)")

figure.savefig('PS1_Q1.png')
```

## Q3.1

Write a function that computes $f(n)=\sum_{i=1}^n 1/(i+1)$. 

```python
def func(n):
    result = 0
    for i in np.arange(1, n+1):
        result += (1/(1+i))
    return result
```

## Q3.2
Plot function $f(n)$ from $n=1$ to $100$.


```python
x = np.arange(1, 101)
y = np.array([])
for i in x:
    y = np.append(y, func(i))
    
figure, axis = plt.subplots()

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Func(x)")
plt.title("Plot of Func(x) function")

figure.savefig('PS1_Q3_2.png') 
```

## Q4 
Plot a histogram of $sin(x)$ where $x$ is chosen from a standard normal distribution (mean = 0, sd = 1). Draw 10,000 samples to make the histogram.

```python
mu, sigma = 0, 1 # mean and standard deviation
x = np.random.normal(mu, sigma, 10000)
y = np.sin(x)

figure, axis = plt.subplots()

plt.hist(y, bins = "auto");
plt.xlabel("X")
plt.ylabel("Sin(x)")
plt.title("Sin(x) from Standard Normal Distribution");

figure.savefig('PS1_Q4.png') 
```

## Q5
Plot a histogram of $exp(x)$ where $x$ is chosen from a uniform distribution in $[0,1.5]$. Draw 10,000 samples to make the histogram.

## Q6.1
Save a numpy array of 1000 random samples from a standard normal distribution (mean = 0, standard deviation = 1). Sort the samples, and plot a histogram of the differences between the adjacent sorted elements (the difference between the first and the second, second and the third, third and the fourth, etc.).

```python
mu, sigma = 0, 1
x = np.random.normal(mu, sigma, 1000)
sorted_x = np.sort(x)

def adjacent_difference(x):
    y = np.array([])
    i = 0
    while i < len(x)-1:
        y = np.append(y, (x[i+1] - x[i]))
        i += 1
    return y

figure, axis = plt.subplots()

data = adjacent_difference(sorted_x)
plt.hist(data, bins=[0, 0.005, 0.01, 0.015, 0.02, 0.025, 0.03]);
plt.xlabel("Difference")
plt.ylabel("Number of Samples")
plt.title("Difference between the adjacent sorted elements");

figure.savefig('PS1_Q6_1.png') 
```

## Q6.2
Plot the difference computed in 6.1 vs. the position in the sorted array (first, second, etc.). 


```python
new_x = np.arange(999)

figure, axis = plt.subplots()

plt.plot(new_x, data);
plt.xlabel("Position of Sorted Array")
plt.ylabel("Difference")
plt.title("Plot of the Difference vs. Sorted Array");

figure.savefig('PS1_Q6_2.png') 
```
